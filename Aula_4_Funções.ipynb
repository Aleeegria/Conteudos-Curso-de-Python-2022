- ### 1.3 Por que funções são importantes?

Imagine que em várias situações eu precise tirar a primeira letra de uma string e colocar no final minúscula




```python
string_1 = 'pato'

print(string_1[2:] + string_1[0:2].upper())
```

```python
string_2 = 'Beatles'

print(string_2[1:] + string_2[0].lower())
```

```python
string_3 = 'Pokémon'

print(string_3[1:] + string_3[0].lower())
```

# <font color=blue>2. Escopo de funções e funções aninhadas</font>

- ### 2.1 Diferença entre variável global e local

```python
a = "variável global"

def func():
    a = "variável local"
    print(a)
    
func()
```

```python
print(a)
```

```python
a = "variável global"

def func():
    global a, b
    a = "Alterando a Variável global dentro da função"
    b = "Outra variável local"
    print(b)
    
func()
```

```python
print(b)
```

# <font color=blue>3. Múltiplos retornos e argumentos opcionais</font>

- ### 3.1 Retornando múltiplos valores

```python
def operacoes(a, b):
    soma = a + b
    diferenca = a - b
    prod = a * b
    div = a / b
    return soma, diferenca, prod, div
```

```python
operacoes(2,3)
```

```python
soma, diferenca, prod, div = operacoes(2,8)

print(f'A soma é de {soma}, a diferença de {diferenca}, o produto é {prod} e o quociente, {div}')
```
